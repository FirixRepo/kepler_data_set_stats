```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


#Please move to /reports/claim-1.md to read to final claim of this project

df = pd.read_csv("../data/raw/kepler.csv")

min_period = 50
df['short_period'] = df['koi_period']<min_period
habitable_planets = df[df['is_habitable']==True]
non_habitable_planets = df[df['is_habitable']==False]

group_a = habitable_planets['short_period'].dropna().values
group_b = non_habitable_planets['short_period'].dropna().values
mean_a =  np.mean(group_a)
mean_b = np.mean(group_b)
diff_obs =mean_a-mean_b

print("Mean of group a ",mean_a)
print("Mean of group b",mean_b)
print("Diff",diff_obs)

all_data = np.concatenate([group_a,group_b])
perm_diffs=[]
n_a = len(habitable_planets)
nb_permutation = 5000

for _ in range(nb_permutation):
    shuffled_data = np.random.permutation(all_data)
    a = shuffled_data[:n_a]
    b = shuffled_data[n_a:]

    diff = np.mean(a)-np.mean(b)
    perm_diffs.append(diff)

perm_diffs = np.array(perm_diffs)

p_value = np.mean(np.abs(perm_diffs)>=np.abs(diff_obs))

print("P_value :",p_value)

#P_value is : 0.0028 so <= to 0.05 suggesting that the difference observed in means
#is unlikely to have occured in the null hypothesis( all planets have the same period)


```
