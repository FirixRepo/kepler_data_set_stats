```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy


#Please move to /reports/claim-4.md to read to final claim of this project

df = pd.read_csv("../data/raw/kepler.csv")

confirmed = df[df['koi_disposition']==2]
not_confirmed = df[df['koi_disposition']!=2]
temp_a = confirmed['koi_teq'].dropna().values
temp_b = not_confirmed['koi_teq'].dropna().values

#Visual
plt.hist(temp_a, bins=30, alpha=0.5, label="CONFIRMED")
plt.hist(temp_b, bins=30, alpha=0.5, label="NOT CONFIRMED")
plt.xlabel("Temperature (koi_teq)")
plt.ylabel("Count")
plt.title("Temperature distribution")
plt.legend()
plt.show()


mean_obs = np.mean(temp_a)-np.mean(temp_b)
print(temp_a.shape)
print(temp_b.shape)
print("Mean observed : ",mean_obs)

#test statistic
t_test,p_value =scipy.stats.ttest_ind(temp_a,temp_b,equal_var=False)
print("t-test statistic : ",t_test)
print("p-value : ",p_value)

```
