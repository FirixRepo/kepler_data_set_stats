```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Please move to /reports/claim-6.md to read to final claim of this project

df = pd.read_csv("../data/raw/kepler.csv")

habitable_planets = df[df['is_habitable']==True]
non_habitable_planets = df[df['is_habitable']==False]
group_a = habitable_planets['koi_prad'].dropna().values
group_b = non_habitable_planets['koi_prad'].dropna().values

n1=len(group_a)
n2=len(group_b)

std_a =np.std(group_a,ddof=1) 
std_b =np.std(group_b,ddof=1)
pooled_std = np.sqrt(((n1-1)*std_a**2 + (n2-1)*std_b**2) / (n1+n2-2))

cohen_d = (np.mean(group_a)-np.mean(group_b))/pooled_std

print("Cohen's d :",cohen_d)
print("Abs cohen's d :",abs(cohen_d))

#Cohen's d : 0.62
#As we see it, even though the result is statistically significant,
#the effect size is moderate meaning that the difference in planet size between habitable and non-habitable planets is meaningful.
```
