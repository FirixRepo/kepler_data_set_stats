```python
import pandas as pd;
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import ttest_ind
from scipy.stats import pearsonr,spearmanr

df = pd.read_csv("../data/raw/kepler.csv")

data = df[['koi_period','koi_prad']].dropna()
data = df[(df['koi_period']>0) & (df['koi_prad']>0)]

radius = data['koi_prad'].values
period = data['koi_period'].values

plt.figure(figsize=(7,5))
plt.scatter(radius,period,alpha=0.4)
plt.xlabel("Planet Radius")
plt.ylabel("Orbital Period")
plt.title("Radius vs Period")
plt.show()

plt.figure(figsize=(7,5))
plt.scatter(radius,period,alpha=0.4)
plt.xscale('log')
plt.ylabel('log')
plt.xlabel("Planet Radius")
plt.ylabel("Orbital Period")
plt.title("Radius vs Period")
plt.show()


pearson_corr,pearson_p = pearsonr(radius,period)

print("Pearson correlation:", pearson_corr)
print("Pearson p-value:", pearson_p)


spearman_corr,spearman_p = spearmanr(radius,period)

print("Spearman correlation",spearman_corr)
print("Spearman p-value",spearman_p)
```
