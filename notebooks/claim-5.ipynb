```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Please move to /reports/claim-5.md to read to final claim of this project

df = pd.read_csv("../data/raw/kepler.csv")

habitable_planets = df[df['is_habitable']==True]
sizes = habitable_planets['koi_prad'].dropna().values
mean_obs = sizes.mean()

print("Habitable planet mean size",mean_obs,"Radii")

nb_boots =1000
boots_means=[]

for _ in range(nb_boots):
    sample = np.random.choice(sizes,size=len(sizes),replace=True)

    boots_means.append(np.mean(sample))

boots_means = np.array(boots_means)

ci_low = np.percentile(boots_means, 2.5)
ci_high = np.percentile(boots_means, 97.5)
print("Difference observed :",mean_obs)
print('IC 95% bootstrap : [',ci_low,',',ci_high,']')

#The mean size of habitable planets is 1.138 Radii 
#With a Confidence interval at 95% of [0.9222884615384617 , 1.3269230769230769 ]
```
